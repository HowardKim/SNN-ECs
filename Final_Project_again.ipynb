```python
import subprocess

n_train = 600
n_test = 50

best_norm = 2 * 34 * 34/10
best_nu1 = 1e-5
best_nu2 = 1e-2
best_exc = 22.5
best_inh = 120

o_norm = [best_norm * (10 ** i) for i in range(-2,2) if i != 0]
o_nu1 = [best_nu1 * (10 ** i) for i in range(-2,2) if i != 0]
o_nu2 = [best_nu2 * (10 ** i) for i in range(-2,2) if i != 0]
o_exc = [best_exc * (2**i) for i in range(-2,2) if i != 0]
o_inh = [best_inh * (2**i) for i in range(-2,2) if i != 0]

commands = ["python",'/content/bindsnet/examples/mnist/nmnist.py', "--gpu", "--n_train", str(600), "--n_test", "50", "--batch_size", "1", "--n_neurons", "100"]
commands += ["--norm", str(best_norm), "--nu1", str(best_nu1), "--nu2", str(best_nu2), "--exc", str(best_exc)]
commands += ["--inh", str(best_inh)]
subprocess.call(commands)

for norm in o_norm:
  commands = ["python",'/content/bindsnet/examples/mnist/nmnist.py', "--gpu", "--n_train", str(600), "--n_test", "50", "--batch_size", "1", "--n_neurons", "100"]
  commands += ["--norm", str(norm), "--nu1", str(best_nu1), "--nu2", str(best_nu2), "--exc", str(best_exc)]
  commands += ["--inh", str(best_inh)]
  subprocess.call(commands)
for nu1 in o_nu1:
  commands = ["python",'/content/bindsnet/examples/mnist/nmnist.py', "--gpu", "--n_train", str(600), "--n_test", "50", "--batch_size", "1", "--n_neurons", "100"]
  commands += ["--norm", str(best_norm), "--nu1", str(nu1), "--nu2", str(best_nu2), "--exc", str(best_exc)]
  commands += ["--inh", str(best_inh)]
  subprocess.call(commands)
for nu2 in o_nu2:
  commands = ["python",'/content/bindsnet/examples/mnist/nmnist.py', "--gpu", "--n_train", str(600), "--n_test", "50", "--batch_size", "1", "--n_neurons", "100"]
  commands += ["--norm", str(best_norm), "--nu1", str(best_nu1), "--nu2", str(nu2), "--exc", str(best_exc)]
  commands += ["--inh", str(best_inh)]
  subprocess.call(commands)

for exc in o_exc:
  commands = ["python",'/content/bindsnet/examples/mnist/nmnist.py', "--gpu", "--n_train", str(600), "--n_test", "50", "--batch_size", "1", "--n_neurons", "100"]
  commands += ["--norm", str(best_norm), "--nu1", str(best_nu1), "--nu2", str(best_nu2), "--exc", str(exc)]
  commands += ["--inh", str(best_inh)]
  subprocess.call(commands)

for inh in o_inh:
  commands = ["python",'/content/bindsnet/examples/mnist/nmnist.py', "--gpu", "--n_train", str(600), "--n_test", "50", "--batch_size", "1", "--n_neurons", "100"]
  commands += ["--norm", str(best_norm), "--nu1", str(best_nu1), "--nu2", str(best_nu2), "--exc", str(best_exc)]
  commands += ["--inh", str(inh)]
  subprocess.call(commands)


```
